```python
import pandas as pd
```

```python
# Read the excel file

data=pd.read_excel('total.xlsx')
data.head()
```

```python
data.columns
```

```python
# Removing Unwanted columns in the table

data.drop(["Unnamed: 0"],axis=1,inplace=True)
data.head()
```

```python
# A function to convert the rating of movie from digits into word such as "Good","Average" & "Bad" 

def change_Rat(r):
    if r>=7:
        return "Good"
    elif r>3.5 or r<7:
        return "Average"
    else:
        return "Bad"
```

```python
# Apply the function on "Rating" column of data

Rating_new=list(map(change_Rat,data["Rating"]))
```

```python
data["Rating"]=Rating_new
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
enc=LabelEncoder()
```

```python
# Encode the string contents in the table

data["Actor_enc"]=enc.fit_transform(data["Actor"])
data["Director_enc"]=enc.fit_transform(data['Director'])
data["Rating_enc"]=enc.fit_transform(data["Rating"])
```

```python
# Alter the NaN value with 0

data.fillna(0,inplace=True)
```

```python
data.head()
```

```python
data.columns
```

```python
# Define the independant (X) features

X=data[['Actor_enc', 'Director_enc','Action', 'Adventure', 'Comedy',
       'Drama', 'Thriller', 'Crime', 'History', 'Mystery', 'Fantasy', 'Family',
       'Romance', 'War', 'Biography']]
X.head(2)
```

```python
# Define the dependant feature 

y=data['Rating_enc']
y.head(3)
```

```python
# Split the data for training and testing

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=.7,random_state=1)
```

```python
# Import the BernoulliNB model

from sklearn.naive_bayes import BernoulliNB
```

```python
model=BernoulliNB()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
# Import the joblib to save the model

import joblib
```

```python
joblib.dump(model,"BernoulliNB_model")
```
